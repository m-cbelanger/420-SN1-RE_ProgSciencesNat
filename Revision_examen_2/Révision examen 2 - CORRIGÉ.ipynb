
### Question 2 (listes associées)

Voici 2 listes associées qui représentent la température d'une piscine en celsius et l'heure de la journée à laquelle la mesure a été prise. Toutes les questions ci-dessous doivent être faite avec du code, sans intervention de l'humain, c'est-à-dire que vous ne pouvez pas intervenir en écrivant des données que vous voyez, ça dois être variable et fonctionner pour d'autres listes dans le futur.


```py
import numpy
temperature = [25.3 , 26.1 , 26.3 , 26.1 , 26.1 , 81.1 , 26.8 , 26.4 , 26.1 , 26.1]
heure_jour  = ["5:00","6:35","7:50","10:02","10:45","12:12","13:15","16:03","17:38","18:20"]
```

a) Quelle est l'étendue (différence entre la plus haute et la plus basse) de température prise dans cette journée? Ça vous semble normal comme réponse?

```py
etendue = max(temperature) - min(temperature)
print(etendue)
```

C'est un très grand écart pour une seule journée... étrange! Voir la suite!

b) Combien de fois avons-nous pris la température 26.1?

Approche de votre choix, on n'a pas besoin de la position

```py
compteur = 0
for element in temperature:
    if element == 26.1:
        compteur = compteur + 1

print(compteur)
```

c) Faire une liste des heures auxquelles les températures étaient à 26.1

Approche range puisqu'on a besoin de la position de la température pour avoir son heure.

```py
heure_voulues = []
for i in range(len(temperature)):
    if temperature[i] == 26.1:
        heure_voulues.append(heure_jour[i])

print(heure_voulues)
```


d) Quelle est la moyenne et l'écart-type de températures?


```py
moyenne = numpy.mean(temperature)
ecart_type = numpy.std(temperature)
```

e) En a) il y a une valeur qui a été entrée en Fahrenheit plutôt qu'en celsius. Trouvez-là à l'aide d'une condition (si la température est au-dessus de (la moyenne + 2 x ecart_types), c'est une valeur qui n'est assurément pas "normale". Trouver une façon sécuritaire **d'enlever** cette valeur et l'heure à laquelle elle a été prise de sorte que votre code soit réutilisable (donc fonctionne pour enlever des valeurs pas normales d'autres listes de températures).

Préalable: j'utilise les variables moyenne et ecart_type de la question d)

```py
temperature_gardees = []

for i in range(len(temperature)):
    if not(temperature[i] > moyenne +2*ecart_type):
        temperature_gardees.append(temperature[i])

print(temperature_gardees)
```

f) Voici la fonction qui transforme les degrés celsius en Fahrenheit:

```py
def conversion(degre):
    fahrenheit = degre * 9/5 + 32
    return fahrenheit
```

Faire une boucle qui parcoure la liste de température et qui **remplace** la valeur dans la liste par la valeur en Fahrenheit en utilisant la fonction. 

```py
for i in range(len(temperature)):
    temperature[i] = conversion(temperature[i])

print(temperature)
```


### Question 3 (listes mixtes)

Voici une liste contenant plusieurs types de variables (entiers (int), nombres décimaux (float), chaîne de caractères (str)). Elle n'a pas de liste associée et a toujours potentiellement des doublons.


```py
liste_melangee = ["10:40", 8, 9.1, "voiture", "voiture",[5,10,-8.4], 400, -6]
```

> Ne pas oublier que la liste interne n'est pas un des 3 types. Ce qu'elle contient sont des chiffres, mais à l'index 6 de la liste_melangee on a une liste (pas un int, float ou str)

a) Faire 3 liste nouvelles dans laquelle copier distinctement les entiers, les chaînes et les nombres décimaux.

```py
liste_entiers = []
liste_chaines = []
liste_decimaux= []
```

b) Faire une boucle pour faire le tri de cette liste_mixte. 

```py
for element in liste_melangee:
    if isinstance(element, int):
        liste_entiers.append(element)
    elif isinstance(element, float):
        liste_decimaux.append(element)
    elif isinstance(element, str):
        liste_chaines.append(element)

print(liste_entiers)
print(liste_chaines)
print(liste_decimaux)
```

c) Faire un décompte du nombre de chiffres (floats et int) en parcourant la liste_mixte dans une boucle.

```py
nombre_chiffres = 0
for element in liste_melangee:
    if isinstance(element, int) or isinstance(element, float):
        nombre_chiffres = nombre_chiffres + 1
        
print(nombre_chiffres)
```
