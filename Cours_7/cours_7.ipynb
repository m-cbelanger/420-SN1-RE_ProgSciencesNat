Imaginons qu'on ait une liste de données sur laquelle on veut faire les mêmes opérations. C'est alors qu'on aura besoin des structures de répétition (boucles). 


Si on veut voir chaque élément d'une liste, on pourrait faire ceci:

```py
ma_liste=[1,2,"abc","salut",62]
print(ma_liste[0])
print(ma_liste[1])
print(ma_liste[2])
print(ma_liste[3])
print(ma_liste[4])
```
C'est pratique pour les listes dont on connaît le nombre d'éléments et si la liste est très courte... la structure de boucle est faite exprès pour cela!

### Boucles

Une boucle est une instruction qui permet de faire une répétition du même traitement sur un intervalle de valeurs. On explorera 2 types principaux:

`Boucle bornée`
Quand on sait combien de fois doit avoir lieu la répétition, on utilise généralement une boucle `for`. L'intervalle est fermé.

`Boucle non bornée`
Si on ne connaît pas à l’avance le nombre de répétitions, on choisit une boucle `while`. L'intervalle est ouvert (on connaît le début, mais pas la fin).


### Boucle `for` (bornée)

Pour parcourir chaque élément de la liste, on utilisera le mot clé `for` `element` `in` `liste`. La variable `element` est une variable dont on choisit le nom. Dans ce format, la variable `element` représente le contenu dans la case en cours dans la liste.

L’instruction `for` est une instruction dont l’en-tête se termine par deux points :, suivie d’un bloc `indenté` qui constitue le corps de la boucle.

On dit que l’on réalise une `itération` de la boucle à chaque fois que le corps de la boucle est exécuté.

```py
for element in [1,2,"abc","salut",62]:
    print("l'élément est: ", element)  # corps de la boucle
```

ou bien:

```py
for element in ma_liste:
    print("l'élément est: ", element)
```

Exemple avec plusieurs traitements dans le corps de la boucle:
```py
somme = 0
for element in [1,6,5,1,2,3,8]:
    if element <= 5:    #début du corps de la boucle
        somme = somme + element
    else:
        somme = somme + 2*element    #fin du corps de la boucle
        
print(somme)
```


### Boucle avec range()

Lorsqu'on fait une boucle avec un `range` plutôt qu'une liste directement, la variable d'itération prend l'**index** de l'élément où il est rendu. Par convention, on appelle ce compteur `i` en programmation.

```py
for i in range(7):
    print("i a pour valeur", i)
```

Si on veut parcourir chaque élément d'une liste en utilisant `i` comme un compteur (de 0 jusqu'au nombre d'éléments dans la liste), on doit utiliser la fonction `len()` et y mettre la liste dedans (la liste ou bien le nom de la variable qui la représente). 

```py
somme = 0
for i in range(len([1,6,5,1,2,3,8])):
    if i <= 5:
        somme = somme + i
    else:
        somme = somme + 2*i
print(somme)
```
Avec la liste dans une variable:
```py
somme = 0
la_liste = [1,6,5,1,2,3,8]

for i in range(len(la_liste)):
    if i <= 5:
        somme = somme + i
    else:
        somme = somme + 2*i
print(somme)
```

Dans le corps de la boucle, on peut faire plusieurs traitements, comme dans n'importe quel code qu'on a vu jusqu'à présent. 
Exemple:

```py
somme = 0    #variable globale, initialisée avec une quantité de 0.

for i in range(4):
    somme = somme + i
    print("la variable i vaut:", i)
    print("la somme vaut maintenant: ",somme)

```

> Note: pour éviter la confusion, on peut nommer la variable `i` pour les boucle avec range et la variable `element` pour les boucles avec in.
