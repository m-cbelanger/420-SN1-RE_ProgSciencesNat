```py
pourcentage_taxes = 14.975

def taxes(montant_avant_taxes, rabais):
    #On aurait pu faire autrement (soustraire le rabais ou calculer les taxes et l'additionner au total)
    montant = (rabais/100 + 1) * montant_avant_taxes * (1 + pourcentage_taxes/100)
    return montant

# montant après taxes pour 150 dollars d'achat et 12% de rabais
print(taxes(150,12))

```
