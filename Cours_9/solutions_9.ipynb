
## Question 1

Afficher le graphique des fonctions théoriques suivantes:

a) $ f(x) = 3x^2 - 19 $ en vert en vous assurant que les axes passent par (0,0) et sont en caractère gras et bleus. Le tracé de la courbe doit être en pointillé avec les bouts ronds. Mettre le label "f(x) = 3x^2 - 19" dans la légende. Mettre le titre suivant: "Une superbe fonction" et des titres aux axes (x et y). Les axes doivent aller de -5 à 5.

- Réponse:

![Question1a.PNG](img/Question1a.PNG)

- Code:

```py
def f(x):
    return 3*x**2 - 19

graph.figure()
valeurs_x= numpy.linspace(-5,5)
valeurs_y = f(valeurs_x)

graph.plot(valeurs_x, valeurs_y, label='f(x)=3x^2 - 19', color = 'green', linestyle = '-', solid_joinstyle = 'round')

graph.xlabel('x')
graph.ylabel('y')
graph.title('Une superbe fonction')

# Ajout des axes plus foncés passant par (0,0)
graph.axhline(0, color='blue', linewidth=1.2)
graph.axvline(0, color='blue', linewidth=1.2)

graph.show()
```


b) $ g(x) = sin(3x)e^{1/x} $ en orange avec une ligne pleine et des axes qui vont de 1 à 50. On ne peut pas commencer à 0 à cause de la division par 0 dans l'exposant de l'exponentielle.

- Réponse

![Question1b.PNG](img/Question1b.PNG)

- Code

```py
def g(x):
    return numpy.exp(1/x)*numpy.sin(3*x)

graph.figure()

valeurs_x= numpy.linspace(1,50)
valeurs_y = g(valeurs_x)

graph.plot(valeurs_x, valeurs_y, color = 'orange')


graph.xlabel('x')
graph.ylabel('y')
graph.title('Une fonction zigzag')

# Ajout des axes plus foncés passant par (0,0)
graph.axhline(0, color='black', linewidth=1.2)
graph.axvline(0, color='black', linewidth=1.2)

graph.show()
```

c) Variante: important pour se rappeler des anciennes notions.

- calculer la moyenne de consommation d'essence et la mettre dans une variable nommée moyenne.
- calculer le max de consommation d'essence et le mettre dans une variable nommée maxi.
- faire une liste vide nommée couleurs.
- parcourir les consommations d'essence. Si elles sont au-dessus de la moyenne, ajouter la moyenne à la liste couleurs. Sinon, ajouter le maxi à la liste de couleur.  


- Réponse

![Question3c.PNG](img/Question3c.PNG)


- Code

```py
graph.figure()

moyenne = numpy.mean(consommation_essence)
maxi=max(consommation_essence)

couleurs =[]
for element in consommation_essence:
    if element < moyenne:
        couleurs.append(moyenne)
    else:
        couleurs.append(maxi)

# Tracé du nuage de points
graph.scatter(puissance_moteur, consommation_essence, c=couleurs, cmap='jet', marker='D')
graph.xlabel('puissance moteur (chevaux vapeur)')
graph.ylabel("consommation d'essence (L/100km)")
graph.title("Consommation d'essence en fonction \n de la puissance moteur")
graph.grid(True)

# Affichage du graphique
graph.show()

```
