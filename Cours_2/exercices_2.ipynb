# Question 3

Pour cette question, il est très utile de prendre un if imbriqué. Il est possible de le faire autrement, si vous préférez.

En biologie, on peut donner du sang à une personne qui a un sang compatible avec le nôtre. Il existe 4 groupes sanguins: A, B, AB et O. Il existe aussi le Rhésus (+ ou -). Les règle pour pouvoir recevoir du sang sont les suivantes:
- Une personne avec Rhésus + peut recevoir des Rhésus + ou -, peu importe.
- Une personne avec Rhésus - ne peut recevoir que d'un autre -.
- Si le groupe sanguin de la personne contient la lettre A (A, AB), elle peut recevoir d'un autre A.
- Si le groupe sanguin de la personne contient la lettre B (B, AB), elle peut recevoir d'un autre B.
- Si le groupe sanguin de la personne est AB, elle peut recevoir d'un autre AB.
- Les personnes de groupe O ne peuvent recevoir que du O, mais peuvent donner à n'importe quelle autre lettre. 
- À titre indicatif, les personnes qui ont le groupe O- sont les donneurs universels et les personne AB+ sont les receveurs universels.

Construire une fonction nommée transfusion_possible(groupe_receveur, rhesus_receveur, groupe_donneur, rhesus_donneur) qui prend la lettre du groupe sanguin du receveur et du donneur, leur Rhésus respectifs (entrer 'plus' ou 'moins', ne pas mettre de symbols)
